```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("br_ons_energia_armazenada_subsistemas.csv")
```

```python
df['data'] = pd.to_datetime(df['data'])  # Converter a coluna de data para datetime
plt.figure(figsize=(10, 6))
for subsistema in df['subsistema'].unique():
    subset = df[df['subsistema'] == subsistema]
    plt.plot(subset['data'], subset['proporcao_energia_armazenada_verificada'], label=subsistema)

plt.title("Proporo de Energia Armazenada ao Longo do Tempo")
plt.xlabel("Data")
plt.ylabel("Proporo de Energia Armazenada (%)")
plt.legend()
plt.grid()
plt.show()
```
