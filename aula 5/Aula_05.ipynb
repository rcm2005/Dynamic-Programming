```python
import pandas as pd

dados1 = pd.DataFrame({
	'ano': [2021, 2021, 2021, 2021],
'bloco': ["ABCD", "BCDE", "BCDEF","DEFG"],
'ndvi': [0.188, 0.330, 0.111, 0.214],
'area': [20.65, 8.45, 7.87, 21.81],
'ambiente': ["A", "B", "B","D"],
})
dados1
```

```python
dados2 = pd.DataFrame({
	'ano': [2021, 2021, 2021, 2021],
 'bloco': ['ABCD', 'BCDE', 'BCDEF','DEFG'],
 'talhao': [5, 10, 6, 2],
 'corte': [4, 5, 18, 18],
 'TCH': [70.75, 26.56, 95.65, 110.96]
})
dados2
```

# Juntar os dados

```python
df_merged = pd.merge(dados1, dados2, on=['ano', 'bloco'])
df_merged
```

```python
df_merged.info()
```

```python
Exerccios

1. Mostre qual a mdia, mediana, min e mximo do ndvi por bloco.
2. Mostre qual a mdia, mediana, min e mximo do ndvi por ambiente.
3. Mostre qual a mdia, mediana, min e mximo do ndvi por corte.
4. Mostre qual a area total do canavial.
5. Mostre qual a area total do canavial por bloco.
6. Mostre qual a area total do canavial por ambiente.
7. Mostre qual a area total do canavial por corte.
```

```python
# 1. Mostre qual a mdia, mediana, min e mximo do ndvi por bloco.

df_bloco_min = df_merged.groupby('bloco')['ndvi'].min()
df_bloco_max = df_merged.groupby('bloco')['ndvi'].max()
df_bloco_media = df_merged.groupby('bloco')['ndvi'].mean()
df_bloco_mediana = df_merged.groupby('bloco')['ndvi'].median()

print(df_bloco_min)
print(df_bloco_max)
print(df_bloco_media)
print(df_bloco_mediana)
```

```python
# 2. Mostre qual a mdia, mediana, min e mximo do ndvi por ambiente.

df_ambiente_min = df_merged.groupby('ambiente')['ndvi'].min()
df_ambiente_max = df_merged.groupby('ambiente')['ndvi'].max()
df_ambiente_media = df_merged.groupby('ambiente')['ndvi'].mean()
df_ambiente_mediana = df_merged.groupby('ambiente')['ndvi'].median()

print(df_ambiente_min)
print(df_ambiente_max)
print(df_ambiente_media)
print(df_ambiente_mediana)
```

```python
# rea do canavial

df_area_canavial = df_merged['area'].sum()
df_area_canavial
```
