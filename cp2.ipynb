```python
def registrar_falhas():
    # solicitar numero de maquinas
    num_maquinas = int(input("Digite o número de máquinas: "))

    # Dicionário para armazenar as falhas
    maquinas = {}

    # registrando as falhas
    for i in range(1, num_maquinas + 1):
        maquina = {}
        num_falhas = int(input(f"Digite o número de falhas para a máquina {i}: "))

        # Solicitar o tipo de cada falha e contar a quantidade de cada tipo
        for j in range(1, num_falhas + 1):
            tipo_falha = input(f"Digite o tipo da falha {j} para a máquina {i}: ")
            maquina[tipo_falha] = maquina.get(tipo_falha, 0) + 1

        # Armazenando as falhas da máquina no dicionário de máquinas
        maquinas[f"Máquina {i}"] = maquina

    return maquinas

def imprimir_resumo(maquinas):
    # percorrer os registros de cada máquina e imprimir o resumo das falhas
    for maquina, falhas in maquinas.items():
        print(f"\nResumo de falhas para {maquina}:")
        for tipo, quantidade in falhas.items():
            print(f"Tipo: {tipo}, Quantidade: {quantidade}")

# Registrar as falhas e imprimir o resumo
falhas_maquinas = registrar_falhas()
imprimir_resumo(falhas_maquinas)

registrar_falhas()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Função para calcular a matriz de rotação
def calcular_matriz_rotacao(angulo_graus):
    angulo_radianos = np.radians(angulo_graus)
   
    matriz_rotacao = np.array([[np.cos(angulo_radianos), -np.sin(angulo_radianos)],
                               [np.sin(angulo_radianos), np.cos(angulo_radianos)]])
   
    return matriz_rotacao

# Função para girar o personagem
def girar_personagem(personagem, angulo_graus):
    matriz_rotacao = calcular_matriz_rotacao(angulo_graus)
   
    personagem_girado = np.dot(personagem, matriz_rotacao)
   
    return personagem_girado

# Função para redimensionar o personagem
def redimensionar_personagem(personagem, fator_escala):
    matriz_escala = np.array([[fator_escala, 0],
                               [0, fator_escala]])
   
    personagem_redimensionado = np.dot(personagem, matriz_escala)
   
    return personagem_redimensionado

# Função para exibir o personagem
def exibir_personagem(personagem):
    x = personagem[:, 0]
    y = personagem[:, 1]

    plt.figure()
    plt.plot(x, y, 'b-')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Personagem')
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# Função principal
def main():
    # Definindo o personagem inicial (um quadrado)
    personagem = np.array([[0, 0], [1, 0], [1, 1], [0, 1], [0, 0]])
   
    print("Personagem original:")
    exibir_personagem(personagem)
   
    # Solicitar o ângulo de rotação
    angulo_rotacao = int(input("Digite o ângulo de rotação em graus: "))
   
    # Solicitar o fator de escala
    fator_escala = float(input("Digite o fator de escala: "))
   
    # Girar o personagem e redimensioná-lo
    personagem_girado = girar_personagem(personagem, angulo_rotacao)
    personagem_redimensionado = redimensionar_personagem(personagem_girado, fator_escala)
   
    print("Personagem após rotação e redimensionamento:")
    exibir_personagem(personagem_redimensionado)

# Verificar se este script é executado diretamente
if __name__ == "__main__":
    main()
```
